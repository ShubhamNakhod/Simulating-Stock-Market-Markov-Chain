```python
# Importing libraries
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Defining states
states = ["Up", "Down", "Same"]

# Defining transition matrix
P = np.array([[0.7, 0.2, 0.1],  # From Up
             [0.3, 0.5, 0.2],  # From Down
             [0.4, 0.3, 0.3]])
```

```python
# Defining Markov Chain Simulation Function
def markov_simulation(start_state, steps):
    state_dict = {"Up": 0, "Down": 1, "Same": 2}
    reverse_dict = {0: "Up", 1: "Down", 2: "Same"}
    
    current_state = state_dict[start_state]
    history = [start_state]
    
    for _ in range(steps):
        next_state = np.random.choice([0, 1, 2], p=P[current_state])
        history.append(reverse_dict[next_state])
        current_state = next_state

    return history
```

```python
# Running the Simulation
num_steps = 50
history = markov_simulation("Up", num_steps)

# Plotting results
plt.plot(history, marker="o")
plt.xticks(rotation=90)
plt.title("Markov Chain Stock Market Simulation")
plt.xlabel("Time Step")
plt.ylabel("State")
plt.show()
```
